## Problem 1: Rapidity and the Lorentz-Invariant Phase Space

The quantity  

$ \displaystyle \frac{d^3p}{E} $

is a Lorentz-invariant element of phase space. In high-energy collisions, it is often convenient to express this measure in terms of the particle’s **rapidity** and **transverse momentum**.

The **rapidity** $y$ is defined by  

$ y = \frac{1}{2}\ln\!\left(\frac{E + p_z}{E - p_z}\right), $

where $E$ and $p_z$ are the energy and the longitudinal (beam-axis) component of the particle’s momentum, respectively.

The **transverse momentum** $p_T$ is defined as  

$ p_T = \sqrt{p_x^2 + p_y^2}, $

i.e., the magnitude of the momentum in the plane perpendicular to the beam direction.

### (a) Derivation of $E$ and $p_z$ in Terms of $y$, $p_T$, and $m$

#### **Step 1: Key Relations**

$$
E^2 = p_T^2 + p_z^2 + m^2
$$

$$
y = \frac{1}{2} \ln \left( \frac{E + p_z}{E - p_z} \right)

$$
p_T = \sqrt{p_x^2 + p_y^2}

---

#### **Step 2: Manipulating the Rapidity Definition**

$$
2y = \ln \left( \frac{E + p_z}{E - p_z} \right)
$$

$$
e^{2y} = \frac{E + p_z}{E - p_z}
$$

$$
E e^{2y} - p_z e^{2y} = E + p_z
$$

$$
E (e^{2y} - 1) = p_z (e^{2y} + 1)
$$

$$
\frac{E}{p_z} = \frac{e^{2y} + 1}{e^{2y} - 1}
$$

Now, using exponentials and hyperbolic functions:
$$
E = e^y (e^y + e^{-y})
$$

$$
p_z = e^y (e^y - e^{-y})
$$

$$
E = 2\cosh(y)
$$

$$
p_z = 2\sinh(y)
$$

$$
\frac{E}{p_z} = \coth(y)
$$

$$
p_z = E \tanh(y)

#### **Step 3: Relate $(E-p_z)(E+p_z)$ to $p_T$ and $m$**

$$
E^2 = p_T^2 + p_z^2 + m^2
$$

$$
E^2 - p_z^2 = p_T^2 + m^2
$$

$$
(E - p_z)(E + p_z) = p_T^2 + m^2
$$

Taking the logarithm:
$$
\ln\left\{ (E - p_z)(E + p_z) \right\} = \ln(p_T^2 + m^2)
$$

Express as:
$$
\ln\left\{ \frac{E + p_z}{E - p_z} \cdot (E - p_z)^2 \right\} = \ln(p_T^2 + m^2)
$$

$$
\ln\left( \frac{E + p_z}{E - p_z} \right) + 2\ln(E - p_z) = \ln(p_T^2 + m^2)
$$

Recall $\ln\left( \frac{E + p_z}{E - p_z} \right) = 2y$:
$$
2y + 2\ln(E - p_z) = \ln(p_T^2 + m^2)
$$

$$
2\ln(E - p_z) = \ln(p_T^2 + m^2) - 2y
$$

$$
\ln(E - p_z) = \frac{1}{2}\ln(p_T^2 + m^2) - y
$$

Exponentiate:
$$
E - p_z = \exp\left( \frac{1}{2}\ln(p_T^2 + m^2) - y \right)
$$

$$
E - p_z = e^{-y} \sqrt{p_T^2 + m^2}
$$

$$
E = e^{-y} \sqrt{p_T^2 + m^2} + p_z
$$

---

#### **Step 4: Expressing $E$ and $p_z$ in Desired Variables**

Start with:
$$
E = e^{-y}\sqrt{p_T^2 + m^2} + E\tanh(y)
$$

$$
E(1 - \tanh(y)) = e^{-y}\sqrt{p_T^2 + m^2}
$$

$$
E = \frac{e^{-y}\sqrt{p_T^2 + m^2}}{1 - \tanh(y)}
$$

Now, substitute the exponential definition for $\tanh(y)$:
$$
= \frac{e^{-y}\sqrt{p_T^2 + m^2}}{1 - \left(\frac{e^{y} - e^{-y}}{e^{y} + e^{-y}}\right)}
$$

$$
= \frac{e^{-y}\sqrt{p_T^2 + m^2}}{\frac{e^{y} + e^{-y} - (e^{y} - e^{-y})}{e^{y} + e^{-y}}}
$$

$$
= \frac{e^{-y}\sqrt{p_T^2 + m^2}}{\frac{2e^{-y}}{e^{y} + e^{-y}}}
$$

$$
= \sqrt{p_T^2 + m^2} \frac{e^{y} + e^{-y}}{2}
$$

$$
= \sqrt{p_T^2 + m^2}\cosh(y)
$$

---

#### **Step 5: For $p_z$**

$$
p_z = E\tanh(y)
$$

$$
= \sqrt{p_T^2 + m^2}\cosh(y)\tanh(y)
$$

$$
= \sqrt{p_T^2 + m^2}\sinh(y)
$$

---

#### **Final Boxed Result**

$$
\boxed{
E = \sqrt{p_T^2 + m^2}\cosh(y)
\qquad
p_z = \sqrt{p_T^2 + m^2}\sinh(y)

### (b) Changing Variables from $p_z$ to $y$ in the Phase Space Volume

### (c) Massless Limit and Pseudorapidity

In the massless limit $m \to 0$, the rapidity $y$ reduces to the pseudorapidity $\eta$.

Let us relate $\eta$ to the polar angle $\theta$ measured from the beam axis.

---

#### Step 1: Express Transverse and Longitudinal Momenta in Spherical Coordinates

$$
p_T^2 = p_x^2 + p_y^2
$$

$$
= p^2 \cos^2{\phi} \sin^2{\theta} + p^2 \sin^2{\phi} \sin^2{\theta}
$$

$$
p_T^2 = p^2 \sin^2{\theta}
$$

$$
p_z^2 = p^2 \cos^2{\theta}
$$

So

$$
p_T = p \sin{\theta}
$$

$$
p_z = p \cos{\theta}
$$

---

#### Step 2: Massless Limit ($m \to 0$)

In this case, $E = p$.

---

#### Step 3: Write Rapidity Formula

$$
y = \frac{1}{2} \ln \left( \frac{E + p_z}{E - p_z} \right)
$$

Substitute $E = p$ and $p_z = p \cos\theta$:

$$
y = \frac{1}{2} \ln \left( \frac{p + p \cos{\theta}}{p - p \cos{\theta}} \right)
$$

$$
= \frac{1}{2} \ln \left( \frac{1 + \cos{\theta}}{1 - \cos{\theta}} \right)
$$

---

#### Step 4: Reduce to a Function of $\theta$

Note that

$$
1 + \cos{\theta} = 2 \cos^2 \left( \frac{\theta}{2} \right)
$$

$$
1 - \cos{\theta} = 2 \sin^2 \left( \frac{\theta}{2} \right)
$$

So

$$
y = \frac{1}{2} \ln \left( \frac{2 \cos^2\left(\frac{\theta}{2}\right)}{2 \sin^2\left(\frac{\theta}{2}\right)} \right)
$$

$$
= \frac{1}{2} \ln \left( \frac{ \cos^2\left(\frac{\theta}{2}\right)}{ \sin^2\left(\frac{\theta}{2}\right)} \right)
$$

$$
= \frac{1}{2} \left( 2\ln \cos\left(\frac{\theta}{2}\right) - 2\ln \sin\left(\frac{\theta}{2}\right) \right)
$$

$$
= \ln \left( \frac{\cos\left(\frac{\theta}{2}\right)}{ \sin\left(\frac{\theta}{2}\right) } \right)
$$

$$
= \ln \left( \cot \frac{\theta}{2} \right)
$$

$$
= -\ln\left( \tan \frac{\theta}{2} \right)
$$

---

#### Final Boxed Result:

$$
\boxed{ \eta = -\ln\left( \tan\frac{\theta}{2} \right) }
$$

where $\eta \equiv y$ in the $m \to 0$ limit.

## Problem 2: Collider Coordinate System

The figure below illustrates the **Collider Detector Coordinate System**. The incoming beam axis is typically labeled as the **z-axis**, which defines the direction of the particle beams in a collider. The **x-axis** and **y-axis** are defined to be perpendicular to the z-axis, forming what is known as the **transverse plane**.

The right-pointing red arrow represents the **final-state particle**, which is produced from the collision. The projection of the final-state particle’s momentum onto the transverse plane is called the **transverse momentum**, denoted as $p_T$.

The angle between the final-state particle’s trajectory and the beam line (z-axis) is called the **polar angle**, denoted as $\theta$. The angle between the transverse momentum $p_T$ and the x-axis is called the **azimuthal angle**, denoted as $\phi$.

As shown in the previous problem, the polar angle $\theta$ has a relationship with the **pseudorapidity**, $\eta$. In collider physics, $\eta$ is often used instead of $\theta$ because pseudorapidity is **Lorentz invariant**, making it a more convenient quantity for describing particle kinematics in high-energy collisions.
