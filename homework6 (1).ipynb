For the process $$ e^+e^-\rightarrow \mu^+\mu^-$$
In the center of mass frame the cross section for unpolarized scattering the cross section is:
   $$                                                                         
   \frac{d\sigma}{d\Omega} = \frac{\alpha^2}{4s} \left (1+\cos^2\theta \right )                                                                        
   $$

where $\alpha = e^2/4\pi $ is the fine structure constant ($\frac{1}{137}$) and $\theta$ is the angle the outgoing $\mu^-$ makes with the beamline.

In this problem, we will continue to explore the process $e^+e^-\rightarrow \mu^+ \mu^-$

### 3a.

If the electron beams are unpolarised, the matrix element cannot depend on $\phi$, since the problem is azymuthally symmetric.   Therefore $d\sigma/d\Omega$ depends only on $\cos\theta$.   Integrate the cross section given earlier over $\phi$ and then make a plot of the predicted angular distribution $d\sigma /d\cos\theta$.

#### $e^+ e^- \rightarrow \mu^+ \mu^-$

---

The differential cross section:

$$
\frac{d\sigma}{d\Omega} = \frac{\alpha^2}{4s} \left(1 + \cos\theta\right)
$$

---

The infinitesimal cross section:

$$
d\sigma = \frac{\alpha^2}{4s} (1 + \cos\theta) d\Omega
$$

Integrate to get the total cross section:

$$
\sigma = \frac{\alpha^2}{4s} \int_0^{2\pi} d\phi \int_0^{\pi} (1 + \cos\theta) \sin\theta \, d\theta
$$

$$
= \frac{\alpha^2 \pi}{2s} \left[ \int_0^{\pi} \sin\theta \, d\theta + \int_0^{\pi} \sin\theta \cos\theta \, d\theta \right]
$$

Let $u = \cos\theta \implies du = -\sin\theta d\theta \implies d\theta = \frac{du}{-\sin\theta}$

So:

$$
= \frac{\alpha^2 \pi}{2s} \left\{ \left[ -\cos\theta \right]_0^\pi + \left[ \frac{\cos^2\theta}{2} \right]_0^\pi \right\}
$$

$$
= \frac{\alpha^2 \pi}{2s} \left\{ \frac{\cos^2\theta}{2} - \cos\theta + \frac{1}{2} \right\}
$$

So as a function of $\cos\theta$:

$$
\sigma(\cos\theta) = \frac{\alpha^2 \pi}{4s} \left\{ \cos^2\theta - 2\cos\theta + 1 \right\}
$$

For the differential with respect to $\cos\theta$:

$$
\frac{d\sigma}{d(\cos\theta)} = \frac{\alpha^2 \pi}{4s} \left\{ 2\cos\theta - 2 \right\}
$$

Or simplified:

$$
\frac{d\sigma}{d(\cos\theta)} = \frac{\alpha^2 \pi}{2s} \left\{ \cos\theta - 1 \right\}
$$

---

```python
import numpy as np
import matplotlib.pyplot as plt

## plotting from CHATGPT

# Constants
alpha = 1/137    # Fine-structure constant
s = 1000         # Example value for Mandelstam variable s (adjust as needed)

# Function for differential cross section
def d_sigma_d_cos_theta(cos_theta):
    return (alpha**2 * np.pi) / (2 * s) * (cos_theta - 1)

# Range for cos(theta)
cos_theta_values = np.linspace(-1, 1, 400)
d_sigma_values = d_sigma_d_cos_theta(cos_theta_values)

# Plot
plt.figure(figsize=(8,5))
plt.plot(cos_theta_values, d_sigma_values, label=r'$\frac{d\sigma}{d(\cos\theta)}$')
plt.xlabel(r'$\cos\theta$', fontsize=14)
plt.ylabel(r'$\frac{d\sigma}{d(\cos\theta)}$', fontsize=14)
plt.title(r'$\frac{d\sigma}{d(\cos\theta)} = \frac{\alpha^2 \pi}{2s} (\cos\theta - 1)$')
plt.axhline(0, color='k', lw=1, ls='--')
plt.legend()
plt.grid(True)
plt.show()
```

### 3b. 

Now integrate over $\cos\theta$ to find the total cross section as function of $s$ in units of nb

Start from:
$$
\sigma(s) = \frac{\alpha^2\pi}{s}\;\mathrm{GeV}^{-2}
$$

Use:
$$
1~\mathrm{GeV}^{-2} = 0.389~\mathrm{mb}
$$

So,
$$
\sigma(s) = \frac{\alpha^2\pi}{s} \times 0.389~\mathrm{mb}
$$

Plug in $\alpha = \frac{1}{137}$:
$$
\sigma(s) = \frac{1}{s} \left(\frac{1}{137}\right)^2 \pi \times 0.389~\mathrm{mb}
$$

Compute the numerical constant:
\[
\left(\frac{1}{137}\right)^2 \approx 5.328 \times 10^{-5}
\]
\[
\pi \times 0.389 \approx 1.221
\]
\[
5.328 \times 10^{-5} \times 1.221 \approx 6.504 \times 10^{-5}
\]

So,
\[
\sigma(s) = \frac{6.504 \times 10^{-5}}{s}~\mathrm{mb}
\]

Convert mb to nb:
$$
1~\mathrm{mb} = 10^3~\mathrm{nb}
$$

$$
\sigma(s) = \frac{6.504 \times 10^{-5} \times 10^3}{s}~\mathrm{nb} = \frac{0.065}{s}~\mathrm{nb}
$$

The file mumu.dat contains data corresponding to an integrated luminosity of 74.674 pb$^{-1}$ collected by the BaBar experiment at the SLAC B-factory.  The file contains events from the process $e^+e^-\rightarrow \mu^+\mu^-(\gamma)$.  (where ($\gamma$) means that the event selection may allow a low-energy photon). The photon emission can be treated as a small radiative correction, which does not modify the gross properties of the $e^+e^-\rightarrow \mu^+\mu^-$ process.  The data were collected at a center-of-mass energy of10.539 GeV. Note, that this is below the B B̄ production threshold.

The following code reads these data file and puts the data into a form that can be easily used in python. (For people who prefer to use root, a mumu.root file is available in this directory)

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Parse the input file.
file = "mumu.dat"

# Inforamtion to offset needed to nterpret the info
# The order of variables is explained in the metadata at the top of the file
# The inital e- direction is the +z direction
#
#  isBCMuMu -- a boolean bit which provides a tighter (but cleaner) selection of di-muons.  You should require this bit to be 1
#  p1Mag -- Magnitude of the momentum of the mu- (in GeV)
#  p1CosTheta -- Cosine of thepolar angle of the mu-track  
#  p1Phi -- Azimuthal angle of the mu- track
#  p1EmcCandEnergy -- Electromagnetic Calorimeter energy associated with the highest-momentum track. For muons, this is expected to be non-zero, but small (<1 GeV)
#  p2Mag -- Magnitude of the momentum of the mu+ track (in GeV)
#  p2CosTheta -- Cosine of thepolar angle of the mu+ track
#  p2Phi -- Azimuthal angle of the mu+ track
#  p2EmcCandEnergy -- Electromagnetic Calorimeter energy associated with the second highest-momentum track. For muons, this is expected to be non-zero, but small (<1 GeV)
#
inMeta = False
isBCMuMu = []
p1Mag = []
p1CosTheta = []
p1Phi = []
p1EmcCandEnergy = []
p2Mag = []
p2CosTheta = []
p2Phi = []
p2EmcCandEnergy = []

inMeta = True
for line in open(file,"r"):
    line = line.strip()
    info = line.split(",")    
    if inMeta and ("<metadata>" in info[0]):
        inMeta = True
    elif inMeta and ("</metadata>" in info[0]):
        inMeta = False
    elif not inMeta:
        isBCMuMu.append(int(info[0]))
        p1Mag.append(float(info[1]))
        p1CosTheta.append(float(info[2]))
        p1Phi.append(float(info[3]))
        p1EmcCandEnergy.append(float(info[4]))
        p2Mag.append(float(info[5]))
        p2CosTheta.append(float(info[6]))
        p2Phi.append(float(info[7]))
        p2EmcCandEnergy.append(float(info[8]))
print(len(isBCMuMu))
print(len(p1CosTheta))
print(len(p2CosTheta))
```

### we see an asymmetry along the x = 0 axis: the muon had more negative values than positive and vice-versa for the anti-muon

Physicists often characterize such parity violating effects in terms of a "forward-backward asymmetry":
$$
A_{FB} = \left (
\frac{N^{\mu-}_{\cos\theta >0} - N^{\mu-}_{\cos\theta <0}}{N^{\mu-}_{\cos\theta >0} + N^{\mu-}_{\cos\theta <0}}
\right )
$$
calculate $A_{FB}$ and its uncertainty for the BaBar data (note:  Because the data falls into one of two possible categories, the uncertainty follows a binomial distribution)

Fit the $\mu^- \; \cos\theta $ distribution to the form
$$
N = N_0 \left (1+A\cos\theta+B\cos^2\theta \right )
$$
Note:  to avoid "edge effects" due to particles exiting the end of the detector, limit your fit to the range $|\cos\theta|<0.7$.  How large is $A$ relative to $B$?

```python
# script from homework 3

from scipy.optimize import curve_fit
cos1_good_7 = np.array(cos1_good) 
cos1_good_7 = cos1_good_7[np.abs(cos1_good_7) < 0.7]

cos2_good_7 = np.array(cos2_good) 
cos2_good_7 = cos1_good_7[np.abs(cos1_good_7) < 0.7]

A_fb = (N_plus - N_minus) / (N_plus + N_minus)

counts, bin_edges = np.histogram(cos1_good_7, bins=20, range=(-0.8,0.8))
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

def fit_func(cos_vals, N0, A, B):
    return N0 * (1 + A*cos_vals + B*cos_vals**2)

p0 = [np.mean(counts), 0, 1]

sel = counts > 0

popt, pcov = curve_fit(fit_func, bin_centers[sel], counts[sel], p0=p0)

N0_fit, A_fit, B_fit = popt

print(f"N0: {N0_fit}, A: {A_fit}, B: {B_fit}")

plt.hist(cos1_good, bins=20, range=(-0.8, 0.8), color='blue', alpha=0.6, label='Data')
xfit = np.linspace(-0.8, 0.8, 100)
plt.plot(xfit, fit_func(xfit, *popt), 'r-', label='Fit')
plt.xlabel(r'$\cos\theta$ for $\mu^-$')
plt.ylabel('Number of Events')
plt.title(r'Angular Distribution: $\mu^-$')
plt.legend()
plt.show()

print(f"A_fb = {A_fb:.3f}")
```

### 3e. 

We can remove this parity violating term by taking the average of the angular distributions for the $\mu^+$ and the $\mu^-$:
$$
\frac{d<N>}{d\cos \theta} = 0.5 \left ( \frac{dN^{\mu+}}{d\cos\theta} + \frac{dN^{\mu-}}{\cos\theta} \right )
$$
where $N^\pm$ are the number of $\mu^+$ and $\mu^-$ respectively.  Make a histogram of this distribution and fit the angular distribution to the form 
$$
N = N_0 \left  ( 1+B \cos^2 \theta\right )
$$
Once again, limit your fit to $|\cos \theta|<0.7$

Calculate the $\chi^2$ between your histogram and your prediction.  Turn the resulting $\chi^2$ into a fit probability.

```python
bins = 20
hist_range = (-0.7, 0.7)

counts1, bin_edges1 = np.histogram(cos1_good_7, bins=bins, range=hist_range)
counts2, bin_edges2 = np.histogram(cos2_good_7, bins=bins, range=hist_range)

bin_centers = (bin_edges1[:-1] + bin_edges1[1:]) / 2  

counts_avg = 0.5 * (counts1 + counts2)
print(counts_avg)
def fit_func_avg(cos_vals, N0, A):
    return N0 * (1 + A*cos_vals**2)

p0 = [np.mean(counts_avg), 0]

sel = counts_avg > 0

popt, pcov = curve_fit(fit_func_avg, bin_centers[sel], counts_avg[sel], p0=p0)

N0_fit, A_fit = popt

print(f"N0: {N0_fit}, A: {A_fit}")

plt.bar(
    bin_centers, counts_avg,
    width=(bin_edges1[1] - bin_edges1[0]),
    color='blue', alpha=0.6, label='Data average'
)
xfit = np.linspace(-0.8, 0.8, 100)
plt.plot(xfit, fit_func_avg(xfit, *popt), 'r-', label='Fit')
plt.xlabel(r'$\cos\theta$ for $\mu^-$')
plt.ylabel('Number of Events')
plt.title(r'Average Angular Distribution: $\mu^- \mu^+$')
plt.legend()
plt.show()

```
