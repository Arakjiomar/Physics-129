# Problem 1: Invariant Mass in a Two-Body Decay

When a particle decays into two other particles, this is called a **two-body decay**.  
If we regard the decaying particle as an isolated system, then by conservation laws:

- Total energy is conserved.  
- Total momentum is conserved.  

In special relativity, the invariant mass is related to the total four-momentum by

$$
m^2 = E^2 - \vec{p}^{\,2},
$$

where $m$ is the mass, $E$ the total energy, and $\vec{p}$ the total momentum (we set $c=1$).  

**Question:**  
Derive the expression for the invariant mass of the decaying particle in terms of the energies of the two decay products ($E_1$, $E_2$) and the opening angle, $\alpha$, between their momenta. 

---



We start from the invariant:
$$
m^2 = E^2 - p^2, 
\qquad 
\vec P = \vec p_1 + \vec p_2, 
\quad 
E = E_1 + E_2
$$

---

$$
m^2 = (E_1+E_2)^2 - \lvert \vec p_1 + \vec p_2 \rvert^2
$$

$$
m^2 = (E_1^2+E_2^2+2E_1E_2) 
      - \left(p_1^{\,2}+p_2^{\,2}+2\,\vec p_1 \cdot \vec p_2\right)
$$

$$
m^2 = (E_1^2-p_1^{\,2}) + (E_2^2-p_2^{\,2}) 
      + 2\,(E_1E_2 - \vec p_1 \cdot \vec p_2)
$$

$$
E_i^2 - p_i^{\,2} = m_i^2
$$
so
$$
m^2 = m_1^2 + m_2^2 + 2\left(E_1E_2 - |\vec p_1||\vec p_2|\cos\alpha\right)
$$

---

**High-energy (ultrarelativistic) limit** $m_i \ll E_i$:

$$
p_i = \sqrt{E_i^2 - m_i^2} 
     = E_i \sqrt{1 - (m_i/E_i)^2} 
     \approx E_i
$$

Therefore:
$$
m^2 \;\approx\; 2E_1E_2(1 - \cos\alpha), 
\qquad (m_i \ll E_i)
$$

# Problem 3

Consider a muon with energy **10 GeV** traversing the sensor.

**Questions:**
1. How much energy does the muon deposit while traversing the 300 μm of silicon?  
2. How many electrons are collected by the detector as a result of this energy deposition?  

---

## Hints for students
- The average stopping power $\langle dE/dx \rangle$ for a minimum-ionizing particle (MIP) in silicon can be found in Particle Data Group\'s website.  
- You may also need to use the density of silicon, which is $\rho \approx 2.33~\text{g/cm}^3$.  
- The average energy required to create one electron-hole pair in silicon is about **3.6 eV**.  

---



## 1

Given:
- Silicon sensor thickness:
$$
w = 300\,\mu\text{m} = 300 \cdot 10^{-4}\,\text{cm} = 0.03\,\text{cm},
$$
- Muon energy:
$$
E_\mu = 10\,\text{GeV},
$$
- Mean energy loss (from PDG):
$$
\left\langle \frac{dE}{dx} \right\rangle_{\text{Si}} = 1.66\,\text{MeV}\,\text{g}^{-1}\text{cm}^2,
$$
- Silicon density:
$$
\rho_{\text{Si}} \approx 2.33\,\text{g}/\text{cm}^3.
$$

From unit analysis, the deposited energy is:
$$
E = \left\langle \frac{dE}{dx} \right\rangle \cdot \rho_{\text{Si}} \cdot w
$$

\
\begin{aligned}
E &= 1.66\,\text{MeV}\cdot\frac{\text{cm}^2}{\text{g}}
    \cdot 2.33\,\frac{\text{g}}{\text{cm}^3}
    \cdot 0.03\,\text{cm} \\[6pt]
  &= 0.1163\,\text{MeV} \\[6pt]
  &= 116{,}300\,\text{eV}.
\end{aligned}

# Problem 4: Position Resolution in a Silicon Strip Detector

A silicon strip detector is used to measure the position of incoming muons.  
The strips have a pitch of **75 μm**. When a muon passes through the detector, it deposits about **$X$ electrons** (you get $X$ from the previous problem) through ionization. These electrons are assumed to be distributed **uniformly** within a horizontal interval of width **75 μm** (±37.5 μm) centered on the muon’s incident position $x_{\text{in}}$.  

The detector strips collect charge in proportion to the overlap between their width and the ionization cloud. The readout system has two possible modes:

---

## 1. Binary Readout
- A strip only registers a hit if it collects more than **600 electrons**.  
- If only one strip registers, the measured position is taken as the **center of that strip**.  
- If two strips register, the measured position is taken as the **midpoint between the two strips** (the boundary).  

**Task:**  
Using a Monte Carlo simulation with hits generated at random $x_{\text{in}}$ uniformly distributed between 0 and 1 cm, estimate the resolution of the measured position (the standard deviation of $x_{\text{measured}} - x_{\text{true}}$).  

---

## 2. Analog Readout (TOT-like)
- Same threshold of **600 electrons** applies.  
- Each strip that passes threshold also records a charge measurement (time over threshold, TOT).  
- The TOT digitization allows **22 discrete values** (0–21), with each step corresponding to **1000 electrons**. For example:  
  - 15,069 electrons → TOT = 15  
  - 23,432 electrons → TOT = 21 (saturation).  
- The reconstructed hit position is:  
  - **Center of the strip** if only one strip fires.  
  - **Charge-weighted centroid** if two strips fire:  
    $$
    x_{\text{rec}} = \frac{Q_1 x_1 + Q_2 x_2}{Q_1 + Q_2},
    $$
    where $Q_i$ are the measured charges after digitization and $x_i$ are the strip centers.  

**Task:**  
Using the same Monte Carlo simulation procedure, estimate the resolution of the analog readout system.  

---

## Overall Goal
- Implement a Monte Carlo simulation where:  
  - Each event generates a random true hit position $x_{\text{in}}$.  
  - The ionization electrons are distributed uniformly in a 75 μm interval around $x_{\text{in}}$.  
  - Charges are collected by strips based on overlap and processed according to the binary or analog readout rules.  
- Compare the resolutions obtained in the two cases.


## Hints:
### Strategy
I would try to do this in four steps:

1. **Truth:** generate random hit positions within a strip region.  
2. **Charge deposit:** calculate the number of electrons collected in each strip from the uniform ionization cloud.  
3. **Measurement:** reconstruct the hit position using binary or analog readout rules.  
4. **Resolution:** compute residuals $(x_\text{measured} - x_\text{true})$ and extract the standard deviation.  

These four steps are analogous to event generation, detector simulation, event reconstruction, and analysis, a typical computation workflow in collider data analysis.

---

### Plots to Produce
To help you understand if your simulation makes sense, produce the following plots as sanity checks:

1. **Truth positions**  
   - Histogram of the generated hit positions.  

2. **Deposited electrons**  
   - Distributions of the collected charge in strip 1 and strip 2.  

3. **Measured vs. truth**  
   - Scatter plot showing reconstructed vs true positions (for binary readout).  

4. **Residuals**  
   - Histograms of $(x_\text{measured} - x_\text{true})$ for binary and analog readout.  
   - Quote the standard deviation of each distribution as the resolution.  


```python
import numpy as np
import matplotlib.pyplot as plt 
## CHAT GPT function for core_resolution as numpy.stdv was not working
def core_resolution(residuals, frac=0.68):
    residuals = np.sort(residuals)
    n = len(residuals)
    k = int(frac * n)  # number of points to keep
    widths = residuals[k:] - residuals[:n-k]  # sliding window
    min_width = np.min(widths)
    return min_width / 2   # half-width = effective σ
def binaryReadout(n_events, binary):
    x_true = []
    x_meas = []
    pitch = 0.0075
    range_e = pitch/2
    total_e = 32306       
    threshold = 600  
    density_e = total_e/pitch
    
    for i in range(n_events):
        x_true.append(np.random.uniform(0,1))
        

    for item in x_true:
        # define the minimum and maximum possible range where the electrons lie 
        cloud_left = item - range_e
        cloud_right = item + range_e

        # find one of the strips in which they lie: for example if strip index = 5.4, set to 5 
        strip_temp = item//pitch ## always get the lower strip 

        # define the boundaries of the strip with lower index 
        strip_low_min = strip_temp*pitch
        strip_low_max = (strip_temp+1)*pitch

        strip_high_min = (strip_temp+1)*pitch
        strip_high_max = (strip_temp+2)*pitch

        # to find the overlap, we take the leftmost point of the cloud and 
        # the left edge of the strip, choose the larger one, then take the rightmost 
        # point of the cloud and the right edge of the strip, choose the smaller one, set overlap to 0

        if cloud_right <= strip_low_min or cloud_left >= strip_low_max:
            overlap_low = 0
        else:
            left_edge = max(cloud_left, strip_low_min)
            right_edge = min(cloud_right, strip_low_max)
            overlap_low = right_edge - left_edge
        
        if cloud_right <= strip_high_min or cloud_left >= strip_high_max:
            overlap_high = 0
        else:
            left_edge = max(cloud_left, strip_high_min)
            right_edge = min(cloud_right, strip_high_max)
            overlap_high = right_edge - left_edge
        # here since the electrons are uniformly distributed, we multiply the total number of electrons by the ratio of the overlap and the pitch (similar to 
        electrons_low = density_e * overlap_low 
        electrons_high = density_e * overlap_high 
        if(binary == True):
            # now for if else statements: if the electrons_low cross the threshold and electons high doesnt, then we go for the lower index, and vice versa, but if both reach it we take the midpoint
            if electrons_low > threshold and electrons_high <= threshold:
                x_meas.append(0.5*(strip_low_min + strip_low_max)) #midpoint of strip
            elif electrons_high > threshold and electrons_low <= threshold:
                x_meas.append(0.5*(strip_high_min + strip_high_max))
            elif electrons_low > threshold and electrons_high > threshold:
                x_meas.append(strip_low_max)
        else:
            TOT_low = min(int(electrons_low // 1000), 21) 
            TOT_high = min(int(electrons_high // 1000), 21)
    
            # now for if else statements: if only one strip fires (TOT>0), take the strip center, else use the charge weighted centroid equation 
            if TOT_low > 0 and TOT_high == 0:
                x_meas.append(0.5*(strip_low_min + strip_low_max)) #midpoint of strip
            elif TOT_high > 0 and TOT_low == 0:
                x_meas.append(0.5*(strip_high_min + strip_high_max))
            elif TOT_low > 0 and TOT_high > 0:
                center_low = 0.5*(strip_low_min + strip_low_max)
                center_high = 0.5*(strip_high_min + strip_high_max)
                x_rec = (TOT_low*center_low + TOT_high*center_high) / (TOT_low + TOT_high)
                x_meas.append(x_rec)

            

    #Plotting with help from CHATGPT
    residuals = np.array(x_meas) - np.array(x_true)
    core_sigma = core_resolution(residuals)


    plt.figure(figsize=(12,4))

    # true hits
    plt.subplot(1,3,1)
    plt.hist(x_true, bins=50, alpha=0.7, color='blue')
    plt.xlabel("True position (cm)")
    plt.ylabel("Counts")
    plt.title("True hit positions")

    # measured hits
    plt.subplot(1,3,2)
    plt.hist(x_meas, bins=50, alpha=0.7, color='green')
    plt.xlabel("Measured position (cm)")
    plt.ylabel("Counts")
    plt.title("Measured hit positions")

    # residuals
    plt.subplot(1,3,3)
    plt.hist(residuals, bins=50, alpha=0.7, color='red')
    plt.xlabel("x_meas - x_true (cm)")
    plt.ylabel("Counts")
    plt.title(f"Residuals (σ = {core_sigma:.4f} cm)")


    plt.tight_layout()
    plt.show()

    return "It worked!"

print("Here is the MC simulation using the binary method")
binaryReadout(1000, True)
print("Here is the MC simulation using the analog method")
binaryReadout(1000, False)

```
