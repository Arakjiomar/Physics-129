## Question 1: Constructing hadrons states using SU(3) raising and lowering operators (25 points)

### Learning objectives
In this question you will:
- Learn how to find the SU(3) wave functions for baryons using SU(3) raising and lowering operators

If we know the SU(3) wave function for one member of an SU(3) multiplet, we can construct the other members of the multiplet using SU(3) raising and lowering operators.  Each multiplet has well defined properties under the interachange of two quarks.  For example, the spin-3/2 baryon decouplet that contains the $\Delta $ is symmetric under interchange, eg $\Delta ^{+} = \frac{1}{\sqrt{3}} \left (uud+udu+duu \right ) $, while the spin-1/2 octet that contains the proton and neutron is  antisymmetric under the interchange of the first two quarks but has no symmetry with respect to the third.

### 1a.

For the spin-3/2 decouplet, start with the $\Omega^{-} = sss$ baryon with strangeness $S=-3$.   Use raising and/or lowering operators to construct the members of the decouplet with $S=-2$, known as the the $\Xi$, and with $S=-1 $, known as the $\Sigma^{*}$.

### 1(a) Constructing the $\Xi$ states from $\Omega^-$ using SU(3) raising operators

We start with the fully strange baryon:
$$
\Omega^- = s s s, \qquad S = -3, \quad s = \tfrac{3}{2}.
$$

---

#### SU(3) generators

The SU(3) raising and lowering operators are built from the Gell–Mann matrices:
$$
\hat{T} = \tfrac{1}{2}\lambda, \qquad
\hat{V}_{\pm} = \tfrac{1}{2}(\lambda_4 \pm i\lambda_5), \qquad
\hat{U}_{\pm} = \tfrac{1}{2}(\lambda_6 \pm i\lambda_7).
$$

They act on the quark flavors as
$$
V_{\pm}: s \leftrightarrow u, \qquad
U_{\pm}: s \leftrightarrow d.
$$

---

#### Acting with $V_{+}$ on $\Omega^-$

Apply the **$V_{+}$** operator to each of the three quarks in $\Omega^-$:
$$
V_{+}\Omega^- = (V_{+1} + V_{+2} + V_{+3})\,sss
= uss + sus + ssu.
$$

The resulting normalized symmetric state is
$$
\boxed{\Xi^{0} = \tfrac{1}{\sqrt{3}}\,(uss + sus + ssu), \qquad S=-2.}
$$

---

#### Acting with $U_{+}$ on $\Omega^-$

Similarly, apply the **$U_{+}$** operator:
$$
U_{+}\Omega^- = (U_{+1} + U_{+2} + U_{+3})\,sss
= dss + sds + ssd.
$$

The resulting normalized symmetric state is
$$
\boxed{\Xi^{-} = \tfrac{1}{\sqrt{3}}\,(dss + sds + ssd), \qquad S=-2.}
$$

---

#### Summary

Using the SU(3) raising operators $V_{+}$ and $U_{+}$ on the seed state $\Omega^-$, we obtain the strangeness $S=-2$ members of the spin-$\tfrac{3}{2}$ decuplet:
$$
\Xi^{0} = \tfrac{1}{\sqrt{3}}\,(uss + sus + ssu),
\qquad
\Xi^{-} = \tfrac{1}{\sqrt{3}}\,(dss + sds + ssd).
$$

---

### Continuing: Raising the $S=-2$ states to obtain the $\Sigma^*$ baryons

We now take the $S=-2$ decuplet members $\Xi^0$ and $\Xi^-$ and act with the SU(3) raising operators
$V_{+}$ and $U_{+}$.

---

#### (i) $V_{+}$ acting on $\Xi^0$

Start from
$$
\Xi^{0} = \tfrac{1}{\sqrt{3}}\,(uss + sus + ssu).
$$

Apply
$$
V_{+}\Xi^{0} = (V_{+1}+V_{+2}+V_{+3})(uss + sus + ssu).
$$

Expanding:
$$
V_{+1}(uss)=uus, \quad V_{+2}(sus)=usu, \quad V_{+3}(ssu)=suu.
$$

Each term appears twice (since $V_{+}$ can act on any of the $s$ quarks in each symmetric term):
$$
2\,uus + 2\,usu + 2\,suu.
$$

Normalize to obtain
$$
\boxed{\Sigma^{+} = \tfrac{1}{\sqrt{3}}\,(uus + usu + suu), \qquad S=-1.}
$$

---

#### (ii) $U_{+}$ acting on $\Xi^{0}$

Now apply
$$
U_{+}\Xi^{0} = (U_{+1}+U_{+2}+U_{+3})(uss + sus + ssu).
$$

Expanding:
$$
U_{+1}(uss)=dus, \quad U_{+2}(sus)=dsu, \quad U_{+3}(ssu)=sdu,
$$
and including all permutations gives
$$
dus + dsu + uds + sdu + usd + sud.
$$

There are six distinct terms (each once), so we normalize by $\sqrt{6}$:
$$
\boxed{\Sigma^{0} = \tfrac{1}{\sqrt{6}}\,(dus + dsu + uds + sdu + usd + sud), \qquad S=-1.}
$$

---

#### (iii) $U_{+}$ acting on $\Xi^{-}$

Start from
$$
\Xi^{-} = \tfrac{1}{\sqrt{3}}\,(dss + sds + ssd).
$$

Apply
$$
U_{+}\Xi^{-} = (U_{+1}+U_{+2}+U_{+3})(dss + sds + ssd).
$$

Expanding:
$$
U_{+1}(dss)=dds,\quad U_{+2}(sds)=dsd,\quad U_{+3}(ssd)=sdd.
$$

Each term appears twice, giving
$$
2\,dds + 2\,dsd + 2\,sdd.
$$

Normalize to obtain
$$
\boxed{\Sigma^{-} = \tfrac{1}{\sqrt{3}}\,(dds + dsd + sdd), \qquad S=-1.}
$$

---

### Summary

By applying $V_{+}$ and $U_{+}$ to the $S=-2$ $\Xi$ states, we obtain the strangeness $S=-1$ members of the spin-$\tfrac{3}{2}$ decuplet:

$$
\Sigma^{+} = \tfrac{1}{\sqrt{3}}(uus+usu+suu),
$$
$$
\Sigma^{0} = \tfrac{1}{\sqrt{6}}(dus+dsu+uds+sdu+usd+sud),
$$
$$
\Sigma^{-} = \tfrac{1}{\sqrt{3}}(dds+dsd+sdd).
$$

### 1(b) Constructing the Kaon States from $\pi^+$

We start from the pion
$$
\pi^+ = -u\bar{d}.
$$

Kaons have strangeness $S = \pm 1$.

---

#### (i) $U_{+}$ acting on $\pi^+$

$$
U_{+}\pi^+ = U_{+}(-u\bar d)
= (U_{+1}+U_{+2})(-u\bar d).
$$

Only the antiquark piece changes:
$$
= 0 + u\bar s = u\bar s.
$$

Charge check:
$$
Q = \tfrac{2}{3} + \tfrac{1}{3} = 1.
$$

Thus
$$
\boxed{K^{+} = u\bar s}.
$$

---

#### (ii) $V_{-}$ acting on $\pi^+$

$$
V_{-}\pi^+ = (V_{-1}+V_{-2})(-u\bar d)
= -s\bar d + 0 = -s\bar d.
$$

Charge check:
$$
Q = -\tfrac{1}{3} + \tfrac{1}{3} = 0.
$$

Thus
$$
\boxed{\overline{K}^{0} = -s\bar d}.
$$

---

#### (iii) $I_{-}$ acting on $\pi^+$

$$
I_{-}\pi^+ = (I_{-1}+I_{-2})(-u\bar d)
= -d\bar d = \pi^0_d.
$$

Now apply $U_{+}$ on $\pi^0_d$:
$$
U_{+}\pi^0_d = (U_{+1}+U_{+2})(-d\bar d)
= -d\bar s.
$$

Charge check:
$$
Q = -\tfrac{1}{3} + \tfrac{1}{3} = 0.
$$

Thus
$$
\boxed{{K}^{0} = -d\bar s}.
$$

---

#### (iv) $I_{-}$ acting on $\pi^+$ (quark part)

$$
I_{-}\pi^+ = u\bar u = \pi^0_u.
$$

Now apply $V_{-}$:
$$
V_{-}\pi^0_u = s\bar u.
$$

Charge check:
$$
Q = -\tfrac{1}{3} - \tfrac{2}{3} = -1.
$$

Thus
$$
\boxed{K^{-} = s\bar u}.
$$

---

### Final Kaon Summary

$$
\boxed{
\begin{aligned}
K^{+} &= u\bar s,\\[4pt]
\overline{K}^{0} &= \frac{\bar d s - s\bar d}{\sqrt{2}},\\[4pt]
{K}^{0} &= \frac{\bar s d - d\bar s}{\sqrt{2}},\\[4pt]

\end{aligned}
}
$$

Under strong interactions, isospin is conserved: 
For $K^- = s\bar{u}$ we have 
$$
Y = +1, \qquad I = \tfrac{1}{2}, \qquad I_3 = -\tfrac{1}{2}.

For the proton $p = uud$ we have 
$$
Y = +1, \qquad I = \tfrac{1}{2}, \qquad I_3 = +\tfrac{1}{2}.

The total third component of isospin is
$$
I_3^{\text{total}} = I_3(K^-) + I_3(p) = -\tfrac{1}{2} + \tfrac{1}{2} = 0.

When combining two particles with $I_1 = I_2 = \tfrac12$, the possible total isospin values follow the addition rule
$$
I_{\text{total}} = |I_1 - I_2|, \ldots , I_1 + I_2
\quad \Longrightarrow \quad
\tfrac12 \otimes \tfrac12 = 0 \oplus 1.
$$

Therefore the $K^-p$ initial state has
$$
\boxed{I_3 = 0 \quad\text{and possible total isospin values } I = 0 \ \text{or}\ 1.}

## Question 3: Hyperfine splitting in the meson multiplets (20 points)

### Learning objectives
In this question you will:
- Use concepts first learned in the context of the hydrogen atom to understand mass splittings between vector and pseudoscalar mesons

The lowest lying pseudoscalar and vector multiplets are both SU(3) nonets (octet plus singlet) with no orbital angular momentum.  The mass splittings within a multiplet come from the difference in mass between the $u$, $d$ and $s$ quarks and to a smaller extent from electromagnetic corrections due to the differences in quark charges.  Mass splitting  between the pseudoscalar and vector multiplets is due to QCD hyperfine splitting.  The origin of this term is the same as hyperfine splitting in hydrogen:  the interaction between the magnetic moments of the two particles in the bound state.

The hyperfine mass splitting for mesons with quark content $q_1 \overline{q_2}$ has the form:
$$ \Delta M ({\rm meson}) = A \frac{{\bf S}_1 \cdot {\bf S}_2}{m_1 m_2} |\psi(0)|^2$$
where $A$ is a constant, ${\bf S}_i$ is the spin of quark $i$, $m_i$ is the mass of quark $i$ and $|\psi(0)|$ is the wave function at the origin.  We cannot calculate $\psi(0)$ from first principles but we can use experimental measurements to determine the combination
$A \left | \psi(0)\right |^2$.
For this problem, use $m_u=m_d=336$ MeV and $m_s=509$ MeV.

### 3a.

By comparing the masses of the $K^0$ and $K^{0 *}$ mesons, estimate the
value of $A |\psi(0)|^2$.


**Spin factors for two spin-$\tfrac12$:**
$$
\mathbf S_1\!\cdot\!\mathbf S_2 =
\begin{cases}
-\tfrac{3}{4}, & S=0 \ \text{(pseudoscalar)}\\[2pt]
+\tfrac{1}{4}, & S=1 \ \text{(vector)}
\end{cases}
$$

**Masses of the spin partners (take unperturbed mass $m_1+m_2$):**
$$
M_P = m_1+m_2 - \frac{3A}{4 m_1 m_2}|\psi(0)|^2, \qquad
M_V = m_1+m_2 + \frac{A}{4 m_1 m_2}|\psi(0)|^2 .
$$

**Vector–pseudoscalar splitting:**
$$
\Delta M \equiv M_V - M_P
= \left(\frac{1}{4} - \left[-\frac{3}{4}\right]\right)\frac{A}{m_1 m_2}|\psi(0)|^2
= \frac{A}{m_1 m_2}|\psi(0)|^2 .
$$

$$
\boxed{\,A|\psi(0)|^2 = \Delta M\, m_1 m_2\, } \qquad (\star)
$$

---

### Apply to the kaons

Quark content (same for both partners): $K^{0},K^{*0} :\ d\bar s$.

Use (constituent) quark masses given in the problem:
$$
m_d = 336~\text{MeV}, \qquad m_s = 509~\text{MeV},
\qquad m_1 m_2 = (336)(509) = 1.71024\times 10^{5}~\text{MeV}^2 .
$$

Experimental masses (central values):
$$
M_{K^{*0}}\approx 896~\text{MeV}, \qquad
M_{K^{0}}\approx 497.6~\text{MeV},
$$
so
$$
\Delta M = M_{K^{*0}}-M_{K^{0}} \approx 398.0~\text{MeV}.
$$

Plug into $(\star)$:
$$
A|\psi(0)|^2 \approx (398.4~\text{MeV})\,(1.71024\times 10^{5}~\text{MeV}^2)
\approx 6.81\times 10^{7}~\text{MeV}^3
\approx 6.8\times 10^{-2}~\text{GeV}^3 .
$$

# Question 4 Determining the $\Omega^-$ Lifetime from Data (30pts)

In this problem, you will use a provided dataset to **extract the lifetime of the $\Omega^-$ baryon**.  

#### Background
- The $\Omega^-$ is a baryon with quark content $sss$.  
- Its **mass** is  
  $$
  m_\Omega = 1.672\ \text{GeV}/c^2.
  $$
- A particle traveling in the laboratory can cover a flight distance $L$ before it decays. This distance is related to the particle momentum measured in the lab frame, the particle's mass, and proper decay time. The proper decay time is the time that elapses between a particle’s creation and decay measured in the particle’s own rest frame.

#### The Data
You are provided with a CSV file:  

**`omega_flight_data.csv`**  
- Column 1: momentum $p$ of the $\Omega^-$ particle, in **GeV/$c$**  
- Column 2: flight distance $L$ of the $\Omega^-$ particle, in **cm**

The file contains 100 simulated $\Omega^-$ decays.  

#### Goal
Report your measurement of Omega particle lifetime.

```python
import numpy as np

# Load CSV (skip header row)
data = np.loadtxt("omega_flight_data.csv", delimiter=",", skiprows=1)

# Columns: momentum (GeV/c), flight distance (cm)
momenta = data[:,0]
distances = data[:,1]

print(momenta[:5], distances[:5])

def decayTime(data):
    momenta = data[:,0]
    distances = data[:,1]
    HBAR_GeVs = 6.582119569e-25 #conversion to secs

    inv_mass = 1.672 #Gev/c^2

    E = np.sqrt(momenta**2 + inv_mass**2)     # GeV
    times = []
    L_GeVinv = []
    for l in distances:
        new_l = l*5.067730716e13
        L_GeVinv.append(new_l)
    betas = []
    gammas = []
    for p in momenta:
        E= np.sqrt(p**2 + inv_mass**2)
        beta = p/E
        gamma = E/(inv_mass)
        betas.append(beta)
        gammas.append(gamma)
        
    for i in range(len(betas)):
        t= ((L_GeVinv[i] / (betas[i])) / gammas[i])*HBAR_GeVs
        print(t)
        
        times.append(t)
        
    avg = np.mean(times)
    return avg

decayTime(data)
```
