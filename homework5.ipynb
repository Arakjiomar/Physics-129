## Question 1: Lorentz Invariant Phase Space (LIPS) (20 points)

 In class we argued that the phase space factor
$$
\frac{d^3p}{E}                                             
 $$
is Lorentz invariant.  Show explicitly that if we move from an initial reference frame $F$ to another frame $F'$ moving with velocity $\beta$ with respect to $F$ that the value of the factor above is in fact Lorentz invariant (you may take the velocity being in n the $z$-direction)

We begin with the Lorentz boost in the $z$ direction:
$$
p'_x = p_x, \qquad
p'_y = p_y, \qquad
p'_z = \gamma (p_z - \beta E), \qquad
E' = \gamma (E - \beta p_z),
$$
where $\gamma = \frac{1}{\sqrt{1-\beta^2}}$.

From the energy–momentum relation
$$
E^2 = m^2 + p_x^2 + p_y^2 + p_z^2
$$
we find, for variations with $p_x,p_y$ fixed,
$$
2E\,dE = 2p_z\,dp_z \quad \Longrightarrow \quad dE = \frac{p_z}{E}\,dp_z.
$$

Taking the differential of $p'_z$:
$$
dp'_z = \gamma (dp_z - \beta\,dE)
       = \gamma\!\left(dp_z - \beta \frac{p_z}{E}\,dp_z \right)
       = \gamma\!\left(1 - \beta \frac{p_z}{E}\right) dp_z.
$$

Hence the transformed volume element becomes
$$
d^3p' = dp'_x\,dp'_y\,dp'_z = dp_x\,dp_y\,\gamma\!\left(1 - \beta \frac{p_z}{E}\right)dp_z.
$$

Divide by $E'=\gamma(E-\beta p_z)$:
$$
\frac{d^3p'}{E'}
= \frac{dp_x\,dp_y\,\gamma\!\left(1 - \beta \frac{p_z}{E}\right)dp_z}
       {\gamma(E - \beta p_z)}
= \frac{dp_x\,dp_y\,dp_z}{E}.
$$

Thus the phase–space measure
$$
\boxed{\frac{d^3p}{E} \;\text{is Lorentz invariant}.}
$$


## Question 2: Solutions to the Dirac Equation (I) Thomson 4.3 (20 points)

    \left (\gamma^\mu p_\mu -m  \right ) u = 0
    $$

    $$
    p_\mu = \left (\begin{array}{c} E\\ p_x \\ p_y \\ p_z \end{array} \right )
    $$

We are using the explicit matrix you wrote for $\gamma^\mu p_\mu$ and simply subtracting $m I_4$:

$$

## Question 3: Properties of $\gamma$ matrices (adapted from Thomson 4.6) (20 points)

Starting with the relationships:
$$
\begin{array}{ccl}
\left ( \gamma^{0} \right )^2 & = & I \\
\left ( \gamma^{i} \right )^2 & = & -I \\
\gamma^{\mu} \gamma^{\nu} + \gamma^{\nu}\gamma^{\mu} & = & 0 \;\;(\mu\ne \nu)\\
\end{array}
$$

(where $I$ is the identity matrix)
show by considering the 3 cases $\mu=\nu =0$, $\mu=\nu \ne 0$ and $\mu\ne \nu$ that

$$
\left \{ \gamma^\mu,\gamma^\nu \right \} = 2g^{\mu \nu}
$$

where $\left \{a,b \right \}\equiv ab+ba$

### Case I: $\mu=\nu=0$
$$
\{\gamma^0,\gamma^0\} = \gamma^0\gamma^0 + \gamma^0\gamma^0 = I + I = 2I = 2g^{00}I.
$$

---
### Case II: $\mu=\nu=i\neq 0$
$$
\{\gamma^i,\gamma^i\} = \gamma^i\gamma^i + \gamma^i\gamma^i = (\gamma^i)^2+(\gamma^i)^2 = -2I = 2g^{ii}I, \quad i\ne 0.
$$

Here
$$
g^{\mu\nu}=
\begin{pmatrix}
1 & 0 & 0 & 0\\[2pt]
0 & -1 & 0 & 0\\[2pt]
0 & 0 & -1 & 0\\[2pt]
0 & 0 & 0 & -1
\end{pmatrix}.
$$

---

### Case III: $\mu\neq\nu$
$$
\{\gamma^\mu,\gamma^\nu\} = \gamma^\mu\gamma^\nu+\gamma^\nu\gamma^\mu = 0,
$$
since by construction $\gamma^\mu\gamma^\nu+\gamma^\nu\gamma^\mu=0$ for $\mu\ne\nu$.

Also $g^{\mu\nu}=0$ when $\mu\ne\nu$, so
$$
\{\gamma^\mu,\gamma^\nu\}=2g^{\mu\nu}I \quad\text{for}\quad \mu\ne\nu.
$$

---

Combining the three cases gives the Clifford algebra relation
$$
\boxed{\{\gamma^\mu,\gamma^\nu\}=2\,g^{\mu\nu}I.}
$$

## Question 4: Helicity  (20 points)

Verify that the helicity operator

$$
\hat h = \frac{\vec\Sigma \cdot \vec p}{2p} = \frac{1}{2p} \left ( 
\begin{array}{cc} \vec \sigma \cdot \vec p & 0 \\ 0 & \vec \sigma \cdot \vec p \end{array}
\right )
$$

commutes with the Dirac Hamiltonian

$$
\hat H = \vec \alpha \cdot \vec p + \beta m
$$

where
$$
\vec \alpha = \left (
\begin{array}{cc}
0 & \vec \sigma \\ \vec \sigma & 0 \\
\end{array}
\right ) \;\;; {\mathrm and} \;\;\;
\beta = \left (
\begin{array}{cc} I & 0 \\ 0 & -I
\end{array}
\right )
$$

We define
$$
\hat h = \frac{1}{2p}
\begin{pmatrix}
\boldsymbol\sigma\!\cdot\!\mathbf p & 0\\[4pt]
0 & \boldsymbol\sigma\!\cdot\!\mathbf p
\end{pmatrix},
\qquad
\hat H =
\begin{pmatrix}
0 & \boldsymbol\sigma\!\cdot\!\mathbf p\\[4pt]
\boldsymbol\sigma\!\cdot\!\mathbf p & 0
\end{pmatrix}
+
\begin{pmatrix}
m & 0\\[4pt]
0 & -m
\end{pmatrix}
=
\begin{pmatrix}
m & \boldsymbol\sigma\!\cdot\!\mathbf p\\[4pt]
\boldsymbol\sigma\!\cdot\!\mathbf p & -m
\end{pmatrix}.
$$

We want to compute the commutator
$$
[\hat h,\hat H]
= \hat h\,\hat H - \hat H\,\hat h.
$$

Explicitly:
$$
[\hat h,\hat H]
=
\frac{1}{2p}
\begin{pmatrix}
\boldsymbol\sigma\!\cdot\!\mathbf p & 0\\[2pt]
0 & \boldsymbol\sigma\!\cdot\!\mathbf p
\end{pmatrix}
\begin{pmatrix}
m & \boldsymbol\sigma\!\cdot\!\mathbf p\\[2pt]
\boldsymbol\sigma\!\cdot\!\mathbf p & -m
\end{pmatrix}
-
\begin{pmatrix}
m & \boldsymbol\sigma\!\cdot\!\mathbf p\\[2pt]
\boldsymbol\sigma\!\cdot\!\mathbf p & -m
\end{pmatrix}
\frac{1}{2p}
\begin{pmatrix}
\boldsymbol\sigma\!\cdot\!\mathbf p & 0\\[2pt]
0 & \boldsymbol\sigma\!\cdot\!\mathbf p
\end{pmatrix}.
$$

Carrying out the block multiplications:
$$
[\hat h,\hat H]
=
\frac{1}{2p}
\begin{pmatrix}
(\boldsymbol\sigma\!\cdot\!\mathbf p)\,m & (\boldsymbol\sigma\!\cdot\!\mathbf p)^2\\[2pt]
(\boldsymbol\sigma\!\cdot\!\mathbf p)^2 & -(\boldsymbol\sigma\!\cdot\!\mathbf p)\,m
\end{pmatrix}
-
\frac{1}{2p}
\begin{pmatrix}
m\,(\boldsymbol\sigma\!\cdot\!\mathbf p) & (\boldsymbol\sigma\!\cdot\!\mathbf p)^2\\[2pt]
(\boldsymbol\sigma\!\cdot\!\mathbf p)^2 & -\,(\boldsymbol\sigma\!\cdot\!\mathbf p)\,m
\end{pmatrix}.
$$

Since $[\boldsymbol\sigma\!\cdot\!\mathbf p, m]=0$ and $(\boldsymbol\sigma\!\cdot\!\mathbf p)^2$ appears identically in both terms, the two matrices cancel exactly. Thus
$$
[\hat h,\hat H] = 0.
$$
